```python
# Include only numbers in the 'price' column
df['price'] = df['price'].replace({'[^0-9]': ''}, regex=True)

# Change empty data to NaN
df['price'] = df['price'].replace('', np.nan)

# Convert data to integer type
df['price'] = df['price'].astype('Int64')
```

```python
df["price"].head()
```

```python
def find_column_outliers(df, column_name):
    column = df[column_name]
    Q1 = np.quantile(column, 0.25)
    Q3 = np.quantile(column, 0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - (1.5 * IQR)
    upper_limit = Q3 + (1.5 * IQR)
    
    outliers = column[(column < lower_limit) | (column > upper_limit)]
    return outliers

```

```python


# find outliers
outliers = find_column_outliers(df, "price")

# print outliers
print(outliers)

# Convert 'price' column to 'float'
df['price'] = df['price'].astype('float')

# calculate upper limit after finding outliers
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)

# outlierları mean ile değiştiriyoruz
df.loc[df['price'] > upper_limit, 'price'] = df['price'].mean()
```
