```python
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd

```

```python

data = pd.read_csv('data.csv', skipinitialspace=True, header=0)
data= pd.DataFrame(data)
my_data = data.iloc[:, 9: 18]
my_data
```

```python
my_data = my_data.applymap(lambda x: x.strip('\r\n'))
my_data = my_data.applymap(lambda x: x.strip(','))
my_data['seller'] = my_data['seller'].apply(lambda x: x.strip('["\n\']').strip('\\').strip('n').split(',')[0].strip("['").replace("['", '') if isinstance(x, str) else x)
my_data['country_version'] = my_data['country_version'].str.split(',').str[1]


my_data.tail(20)
```
